```python
import os
import re
import nibabel as nib
import numpy as np

from glob import glob
import matplotlib.pyplot as plt
```

```python
# !pip install matplotlib
```

```python
def natural_keys(text):
    return [ int(c) if c.isdigit() else c for c in re.split(r'(\d+)', text) ]
```

```python
ROOT_PATH="/U-Net/data/PROSTATE_TASK2/UNCERTAINTY_CLASS/labelsTr"
dirctory_names=glob(f"{ROOT_PATH}/*", recursive = True)
dirctory_names.sort(key=natural_keys)

for index,dir_name in enumerate(dirctory_names):
    # print(index,dir_name)
    img =nib.load(dir_name).get_fdata()
    
    unique_data=np.unique(img,return_counts=True)
    print(dir_name.split("/")[-1],unique_data)
    # os.rename(dir_name,f'{ROOT_PATH}/{index+1}')
    
    
    unique_index = list(map(int, unique_data[0]))
    fig, axs = plt.subplots(1, len(unique_index), figsize=(6, 6))
    
    for index,seg_index in enumerate(unique_index):
        
        ax = axs[index]
        
        masked_seg = np.where(img == seg_index, img, 0)
        ax.imshow(masked_seg, cmap='gray')
        
        # print(np.unique(seg))

        ax.axis('off')
    plt.tight_layout()
    plt.show()
```
